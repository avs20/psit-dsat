```python
# generate a pandas dataframe of age and salary data using pandas 
# and then use linear regression to predict salary based on age

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def gen_data(nrows = 100):
    """
    Generate a pandas dataframe of age and salary data using pandas 
    make sure the salary is a linear function of age
    """
    age = np.random.randint(18, 65, nrows)
    salary = 1000 + 20 * age + np.random.normal(0, 100, nrows)
    data = pd.DataFrame({'age': age, 'salary': salary})
    return data




data = gen_data()
print(data)

```

```python
# plot the data
plt.scatter(data['age'], data['salary'])
plt.xlabel('age')
plt.ylabel('salary')
plt.show()
```

```python
%matplotlib inline

# plot a linear regression line on the data
# use the sklearn library
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(data[['age']], data['salary'])
print(model.coef_)
print(model.intercept_)
print(model.score(data[['age']], data['salary']))
# plot the line 
x = np.arange(0, 70)
y = model.coef_ * x + model.intercept_
plt.plot(x, y, 'r-')
plt.scatter(data['age'], data['salary'])
plt.xlabel('age')
plt.ylabel('salary')
plt.savefig('lr-solution.png', dpi=300)

plt.show()

```

## 1. Do it randomly 

```python
# create a function that returns 100 lines of random slope and intercept
# np.random.seed(50000)
# np.random.seed(None)
def get_lines(n = 10000):
    lines = []
    for i in range(n):
        m = np.random.normal(0, 50, 1)
        b = np.random.normal(0, 100, 1)
        lines.append((m,b))
    return lines


# create a function which calculates residuals for a line and data
def get_residuals(line, data):
    m, b = line
    x = data['age']
    y = data['salary']
    y_pred = m * x + b
    residuals = y - y_pred
    return residuals


# create a function which calculates the sum of squared residuals
def get_ssr(line, data):
    residuals = get_residuals(line, data)
    ssr = np.sum(residuals ** 2)
    return ssr

# create a function to calculate the best line
def get_best_line(data):
    lines = get_lines()
    best_line = None
    best_ssr = None
    for line in lines:
        ssr = get_ssr(line, data)
        if best_ssr is None or ssr < best_ssr:
            best_ssr = ssr
            best_line = line
    return best_line

# get the best line
best_line = get_best_line(data)
print(best_line)

# plot the best_line on data 
m, b = best_line
x = np.arange(0, 70)
y = m * x + b
plt.plot(x, y, 'r-')
plt.scatter(data['age'], data['salary'])
plt.xlabel('age')
plt.ylabel('salary')
plt.show()
```

## 2. Loss function minimization using maths 



```python
def loss_function(w, b, points):
    total_error = 0 

    for i in range(len(points)):
        x = points.iloc[i,0]
        y = points.iloc[i,1]
        total_error  += (y - (w * x + b)) ** 2
    
    return total_error / float(len(points))

losses = []

def gradient_descent(w_now, b_now, points, lr):
    w_gradient = 0 
    b_gradient = 0 

    n = len(points)

    for i in range(n):
        x = points.iloc[i,0]
        y = points.iloc[i,1]


        w_gradient += (-2 / n) * x * ( y- w_now * x +b_now)
        b_gradient += (-2 / n) * ( y- w_now * x +b_now)
    
    w = w_now - w_gradient * lr 
    b = b_now - b_gradient * lr

    losses.append(loss_function(w,b, points))

    return w, b 


w = 100 
b = 0 
lr = 0.0001
epochs = 100

for i in range(epochs):
    w, b = gradient_descent(w, b, data, lr)

print(w, b)

plt.scatter(data['age'], data['salary'], color = 'black')
plt.plot(data['age'], w * data['age'] + b, color = 'blue', linewidth = 3)
plt.show()

```
